```python
import pandas as pd
import spacy

# Example data (replace with actual data source)
data = {
    'Symbol': ['GOOGL', 'AAPL', 'MSFT', 'AMZN'],
    'Company_Name': ['Alphabet Inc.', 'Apple Inc.', 'Microsoft Corporation', 'Amazon.com Inc.'],
    'Industry': ['Technology', 'Technology', 'Technology', 'Retail'],
    'Market Cap (in billions USD)': [1900, 2600, 2000, 1700]
}

# Create DataFrame
df = pd.DataFrame(data)
```

```python
df
```

```python
symbols=df["Symbol"].tolist()
companies=df["Company_Name"].tolist()
```

```python
nlp=spacy.load("en_core_web_sm")

ruler=nlp.add_pipe("entity_ruler",before="ner")

patterns=[]

for symbol in symbols:
    patterns.append({"label":"STOCK","pattern":symbol})
for company in companies:
     patterns.append({"label":"ORG","pattern":company})

ruler.add_patterns(patterns)
```

```python
text="""'Symbol': ['GOOGL', 'AAPL', 'MSFT', 'AMZN'],
    'Company_Name': ['Alphabet Inc.', 'Apple Inc.', 'Microsoft Corporation', 'Amazon.com Inc.'],
    'Industry': ['Technology', 'Technology', 'Technology', 'Retail'],
    'Market Cap (in billions USD)': [1900, 2600, 2000, 1700]"""
```

```python
doc=nlp(text)
d={}
for ent in doc.ents:
  print(ent.text,ent.label_)
  d[ent.text]=ent.label_

```
